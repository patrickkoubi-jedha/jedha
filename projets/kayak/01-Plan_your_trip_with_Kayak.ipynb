### recuperer les infos m√©t√©o

-   utiliser l'API d'openweathermap pour r√©cup√©rer les donn√©es m√©t√©o de chaque ville par coord lat et long


````python
from time import sleep
from numpy import long
import requests
import json



                    ############ API NOMINATIM pour r√©cup√©rer les coordonn√©es gps de chaque ville #######################

"""  INFOS API NOMINATIM (OpenStreetMap)

Pas d‚Äôauthentification / pas de cl√© API n√©cessaire, mais il y a des r√®gles tr√®s strictes √† respecter (sinon l' IP se fait bloquer rapidement).
R√®gles importantes en 2026 (toujours les m√™mes depuis des ann√©es) :

1 requ√™te par seconde maximum (pas plus)
Mettre un User-Agent personnalis√© et identifiable (pas de ¬´ python-requests ¬ª tout seul)
Ajouter Referer ou Contact si possible
Pas de g√©ocodage massif / pas de scraping / pas d‚Äôutilisation en batch sans cache

-->> Toujours afficher l‚Äôattribution ¬´ ¬© OpenStreetMap contributors ¬ª """


# ON VEUT JUSTE LA LATTITUDE ET LA LONGITUDE

#on va devoir boucler sur les 35 villes et faire une requete pour chacune d'entre elles, 
# en respectant le rate-limit de 1 requete par seconde

def get_lat_et_long(ville):
    url = "https://nominatim.openstreetmap.org/search"
    params = {
        "q": ville + ", France",
        "format": "json",      #format=json ou format=jsonv2 (les deux marchent bien)      
        "limit": "1",          #pour n'avoir qu'un seul r√©sultat
        "addressdetails": "0", #pas besoin du d√©tail de l'adresse
        "extratags": "0",      #pas besoin de tags suppl√©mentaires
        "namedetails": "0"     #pas besoin de noms alternatifs
    }
    headers = {
        "User-Agent": "ProjetKayakDataScience/1.0 (patrickkoubi@gmail.com)",
        "Referer": "https://jedha.co"   # optionnel mais appr√©ci√©
    }
    # Important : respecter le rate-limit
    try:
        sleep(1.1)  # ‚Üê au moins 1 seconde entre chaque appel
        r = requests.get(url, params=params, headers=headers)
        r.raise_for_status
        data = r.json()
        if data:
            return  {
                "lat": float(data[0].get("lat")),
                "lon": float(data[0].get("lon"))
            }
    except Exception as e:
        print(f"Exception pour {ville}:", e)
        return None

coord_villes=[ get_lat_et_long(ville) for ville in top_35_villes_fr ]

coord_villes
````

## Pour d√©finir une m√©t√©o agr√©able comme :

-   Peu de pluie
-   Temp√©rature agr√©able
-   Pas trop d'humidit√©
-   Ensoleill√© 
-   pas trop de vent 

on va garder les champs:
-   **rain** sum(rain) < 5 mm sur 7 jours
-   **temp** (min / max / day / night‚Ä¶) moyenne des temp.day entre 18‚Äì28 ¬∞C
-   **humidity** moyenne humidity < 70‚Äì75 %
-   **["weather"][0]["id"]** entre 800‚Äì803 (clear / few clouds) : on peut faire une moyenne de weather.id pour les 7 jours, et garder les villes o√π la moyenne est entre 800 et 803
-   **wind_speed** moyenne wind_speed < 10 m/s

process: liste ville depart-> dictionnaire ville/lat/long --> dictionnaire ville/lat/long/meteo 

````python
import os
from time import sleep
import requests
#on recupere la cle api d'openweathermap dans les variables d'environnement
#Plan gratuit : 60 appels/minute + 1 000 000 appels/mois
OPENWEATHERMAP_API_KEY=os.getenv("OPENWEATHERMAP_API_KEY")

#l'api fourni des reponse intergrant le ressent 'feel_like 
# il int√®gre humidit√© + vent pour le confort humain
url = "https://api.openweathermap.org/data/3.0/onecall"

def get_meteo(lat,lon):
    params = {
        "lat": lat,
        "lon": lon,
        "appid": OPENWEATHERMAP_API_KEY,
        "units": "metric",
        "lang": "fr",
        # On exclut ce qu'on n'utilise pas pour all√©ger la r√©ponse
        "exclude": "minutely,hourly,alerts"
    }
    try:
        sleep(1.1)  # respecter le rate-limit de 60 appels/minute
        r = requests.get(url, params=params)
        r.raise_for_status
        return r.json()
    except Exception as e:
        print(f"Exception pour lat={lat}, lon={lon}:", e)
        return None
    

    
data=get_meteo(48.8566, 2.3522) # test pour Paris
data
````

#### On teste pour une ville donn√©e

````python
#on cr√©e une fonction qui prend les donn√©es meteo brutes pour une ville
# et nous rend les infos m√©t√©o synthetis√©es qu'on a d√©finies dans le cahier des charges
#sous forme de dictionnaire 
# l'id√©e est d'ajouter chaque dictionnaire de ville dans une liste de r√©sultats (qu'on pourra ensuite transformer en dataframe)

def analyse_meteo(ville,data):
    # on recup√®re les dates de debut et fin de la prevision pour pouvoir les passer comme duree du sejour √† booking
    date_debut = data["daily"][0]["dt"]
    date_fin = data["daily"][-1]["dt"]
    #pour chaque ville
    #on boucle sur les jours de la p√©riode (8 jours) 
    # #les donn√©es m√©t√©o sont organis√©es dans une liste sous la cl√© "daily"


    # on filtre sur les jours o√π la cl√© 'rain' existe et est > 0, 
    # et on somme pour avoir le nombre total de jours de pluie
    jours_pluie = sum(1 for jour in data["daily"] if jour.get('rain', 0.0) > 0) 

    # on fait la somme de la pluie sur tous les jours (en consid√©rant 0 mm pour les jours sans pluie) 
    # et on divise par le nombre total de jours pour obtenir la moyenne
    pluie_moyenne = sum(jour.get('rain', 0.0) for jour in data["daily"]) / len(data["daily"]) 

    #on fait la moyenne de la temp√©rature du jour "day" sur tous les jours de la p√©riode
    temp_moyenne = sum([jour["temp"]["day"] for jour in data["daily"]])/ len(data["daily"])

    # Soleil : vraiment ensoleill√© = id == 800 (clear) ou <= 801 (few clouds)
    # A NOTER --> OpenWeather ne met jamais de 'rain' > 0 quand le code principal est 800‚Äì803
    jours_ensoleilles = sum(1 for jour in data["daily"] if 800 <= jour["weather"][0]["id"] <= 803)

    vent_moyen = sum(jour["wind_speed"] for jour in data["daily"]) / len(data["daily"])

    return{
        "date_debut": date_debut,
        "date_fin": date_fin,
        
        "ville": ville,
        
        #on garde aussi les coordonn√©es gps pour pouvoir faire le lien avec les donn√©es d'hotels ensuite
        'lat': data["lat"],
        'lon': data["lon"],

        "jours_pluie": jours_pluie,
        "pluie_moyenne_mm": pluie_moyenne,
        "temp_moyenne_C": temp_moyenne,
        "jours_ensoleilles": jours_ensoleilles,
        "vent_moyen_m_s": vent_moyen
    }
meteo_paris=analyse_meteo("Paris",data)
print(meteo_paris)
````

on va faire une fonction qui prend toutes les villes en entr√©e, et qui pour chaque ville va faire une requete √† l'API d'openweathermap pour r√©cup√©rer les donn√©es m√©t√©o, puis va calculer les indicateurs de m√©t√©o agr√©able d√©finis ci-dessus, et enfin va retourner un dictionnaire avec les r√©sultats pour chaque ville.

````python
def obtenir_meteo_villes(villes):
    resultats = []
    for ville in villes:
        print(f"Traitement de {ville}...")
        coord = get_lat_et_long(ville)
        if coord:
            data_meteo = get_meteo(coord["lat"], coord["lon"])
            if data_meteo:
                meteo_synthetique = analyse_meteo(ville,data_meteo)
                resultats.append(meteo_synthetique)
    return resultats

resultats = obtenir_meteo_villes(top_35_villes_fr)
resultats
````
