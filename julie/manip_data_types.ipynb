```python
#la liste de départ
michel=["Hello", "I", "Michel", "am"]
```

```python
#on affiche chaque element de la liste initiale
def affiche_liste(liste):
    for el in liste:
        print(el)
affiche_liste(michel)
```

```python
#On intervertit le dernier element avec l'avant dernier
michel[-1], michel[-2] = michel[-2], michel[-1]
#on affiche chaque element de la liste modifiée

affiche_liste(michel)
```

```python
age=[12, 16, 34, 58, 9]
```

```python
for i in range(len(age)):
    for j in range(i+1,len(age)):
        if age[i]>age[j]:
            age[i],age[j]=age[j],age[i]

print(age)
```

```python
kilometres_travelled = [50, 10, 100, 25, 1000, 21, 12,30]
```

```python
total_km=0
compteur=0
for km in kilometres_travelled:
    total_km+=km
    compteur+=1
average=total_km/compteur
```

```python
print(f"On average, users travelled {average} km to get to work.")
```

I don't think this average of 156.0 km to get to work is representative because there is on extreme value of 1000 km. it is unrealistic to travel 1000 km every morning togo to work, so this value of 1000 km is almost certainly an error or an exception.We can try calculating the average without the 1000 value : 

```python
total_km = 0
compteur = 0
for km in kilometres_travelled:
    if km <= 200:           # or km < 300, etc.
        total_km += km
        compteur += 1

average = total_km / compteur if compteur > 0 else 0
print(f"On average, users travelled {average} km to get to work.")
```
