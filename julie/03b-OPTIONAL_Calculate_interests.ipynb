# How much do you get for your savings ?

It is said that if Christopher Columbus had set aside €1 that he had left at 2% interest per year, today his heirs would no longer need to work to live their lives. Let's try to calculate the same thing for our savings. 

1. So let's create a function that will allow us to know how much money a user will have in total after a certain number of years.

2. The user should be able to call the function which will then ask for :

    1. The total amount he wishes to place
    2. The number of years he wants to invest his money for
    3. The rate of interest to which he is entitled
    
3. Pay attention to the errors that the user may enter.

    4. If he enters a string of characters rather than a number, an error should be displayed.
    5. If the user enters negative numbers, an error should also be output.
    6. Finally, it is very likely that the user will write the interest rate as a percentage (e.g. 10%). Raise an error if this is the case, to prevent him from inputting a percentage.
    
4. Finally, the program should only show one number with a maximum of two digits after the decimal point.


```python
# function that will allow us to know how much money a user will have in total after a certain number of years
def calculate_amount():
    # we will ask the user for the deposit amount, the number of years and the interest rate. We will also check 
    # if the input is valid and if not we will ask the user to enter a valid input.
    while True:
        try:
            # we will replace the comma with a dot to allow the user to enter the amount in the format of 18452,5 or 18452.5
            deposit=input("Enter the total amount: ").strip().replace(",", ".")
            deposit=float(deposit)
            #we will check if the deposit is a strictly positive number, if not we will raise a ValueError and ask the user to enter a valid input.
            if deposit<=0:
                raise ValueError("Invalid input. Please enter a strictly positive number.")
            break
        # if the user enters an invalid input, we will catch the ValueError and print the error message and ask the user to enter a valid input.
        except ValueError as e:
            print(e)
            print("Please enter a valid deposit amount . e.g.: 18452.5 \n")
    # we will do the same for the number of years and the interest rate, we will check if the input is valid and 
    # if not we will ask the user to enter a valid input.
    while True:
        try:
            # we will check if the user enters a valid number of years, if not we will raise a ValueError and ask the user to enter a valid input.
            years=input("Enter the number of years: ").strip()
            years=int(years)
            # we will check if the number of years is a strictly positive number, if not we will raise a ValueError and ask the user to enter a valid input.
            if years<=0:
                raise ValueError("Invalid input. Please enter a strictly positive number , e.g.: 12.")
            break
        except ValueError as e:
            print(e)
            print("Please enter a valid number of years . e.g.: 12 \n")
    # we will do the same for the interest rate, we will check if the input is valid and if not we will ask the user to enter a valid input.
    while True:
        try:
            rate=input("Enter the interest rate: ").strip()
            # we will check if the user enters a percentage sign, 
            # if yes we will raise a ValueError and ask the user to enter a valid input.
            if "%" in rate:
                raise ValueError("Invalid input. Please do not enter the percentage sign, e.g.: 0.42.")
            rate=float(rate.replace(",", "."))
            if rate<=0 or rate>=1:
                raise ValueError("Invalid input.Please enter a number between 0 and 1, e.g.: 0.42.")
            break
        except ValueError as e:
            print(e)
            print("Please enter a valid interest rate . e.g.: 0.42 \n")
    # we will calculate the total amount of money the user will have after the specified number of years 
    # using the formula: A = P(1 + r)^n, where A is the total amount, P is the initial deposit, r is the interest rate and n is the number of years.
    total_amount=deposit * (1 + rate)**years
    # we will print the total amount of money the user will have after the specified number of years, 
    # we will also format the total amount to 2 decimal places.
    print(f"The total amount of money you will have after you deposit {deposit} at the end of {years} years will be {total_amount:.2f} €")

calculate_amount()
```
