There's a problem with our quizz : if the user has not given the right answer, the program goes to next question. We would rather like to ask again the same question if the answer was wrong.

5. Using a `while` loop, make sure that the same question will be asked again when the answer was wrong

6. Rewrite your code such that this behaviour is applied to all questions.

```python
# 1. Ask 3 questions to a user via the console 
# 2. Verify whether the answer is correct or not 
# 3. If the answer is wrong, ask the same question again until the user gives the right answer
def quizz():
    
    print("Here you have our quiz !")
    print("---------------------------------")
    
    # We can use a dictionary to store the questions and their corresponding answers
    liste_questions={"What is the capital of France ? ": "paris", 
                     "what is the currency of Japan ? ": "yen", 
                     "what is the name of the largest mammal on Earth ? ": "blue whale"}
    # We can use a loop to iterate through the questions and ask them to the user
    answer=""
    for uneQuestion in liste_questions:
        # We can use a while loop to keep asking the same question until the user gives the right answer
        while answer != liste_questions[uneQuestion]:
            # We can use the input function to ask the question to the user and store the answer in a variable
            answer=input(uneQuestion).lower()
            print(f"Your answer : {answer}")
            # We can use an if statement to check if the answer is incorrect and print a message if it is the case
            if answer != liste_questions[uneQuestion]:
                print("Too bad! That is not the correct answer")
        # If the answer is correct, we can print a congratulatory message and move on to the next question
        print("Good job! This is the right answer")
    # After the user has answered all the questions correctly, we can print a final congratulatory message
    print("Well done! You are the quiz winner!")

quizz()
```

Now, we would like to tackle a last problem : we want to allow only 3 wrong answers (among all questions).

7. Improve your code : if the user makes 3 wrong answers, the program should stop and print ("Too bad, you lost the game !")


```python
# 1. Ask 3 questions to a user via the console 
# 2. Verify whether the answer is correct or not 
# 3. If the answer is wrong, ask the same question again until the user gives the right answer
def quizz():
    
    print("Welcome to our quizz!")
    nb_lives = 3
    print(f"You have {nb_lives} lives.")
    
    # We can use a dictionary to store the questions and their corresponding answers
    liste_questions={"What is the capital of France ? ": "paris", 
                     "what is the currency of Japan ? ": "yen", 
                     "what is the name of the largest mammal on Earth ? ": "blue whale"}
    # We can use a loop to iterate through the questions and ask them to the user
    answer=""
    for uneQuestion in liste_questions:
        # We can use a while loop to keep asking the same question until the user gives the right answer
        while answer != liste_questions[uneQuestion]:
            # We can use the input function to ask the question to the user and store the answer in a variable
            answer=input(uneQuestion).lower()
            print(f"Your answer : {answer}")
            if answer != liste_questions[uneQuestion]:
                nb_lives -= 1
                print(f"Sorry, you have {nb_lives} chances left.")
                if nb_lives == 0:
                    print("Oh no, you lost the quiz...")
                    return
        # If the answer is correct, we can print a congratulatory message and move on to the next question
        print("Good job! This is the right answer")
    # After the user has answered all the questions correctly, we can print a final congratulatory message
    print("Well done! You are the quiz winner!")

quizz()
```
