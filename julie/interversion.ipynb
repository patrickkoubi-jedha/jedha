```python
#la liste de départ
liste_mots=["Hello", "I", "Michel", "am"]
#on affiche chaque element de la liste initiale
print("Liste initiale:")
print("------------------")
for el in liste_mots:
    print(el)
#On intervertit le dernier element avec l'avant dernier
liste_mots[-1], liste_mots[-2] = liste_mots[-2], liste_mots[-1]
#on affiche chaque element de la liste modifiée
print()
print("Liste modifiée:")
print("------------------")
for el in liste_mots:
    print(el)
```
