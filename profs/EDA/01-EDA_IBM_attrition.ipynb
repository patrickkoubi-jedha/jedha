## Who are IBM's employees?
We'll first focus on very basic questions about age, gender, education level, etc...

6. How are the ages distributed?

```python
px.histogram(df, x="Age", title="Age distribution", width=700).show()
```

7. Is the distribution of ages different for men and women?
Find two ways of investigating the difference

The distribution of ages is similar for men and women

```python
age_gender_counts = df.groupby(["Age", "Gender"]).size().reset_index(name='count')
fig2=px.bar(age_gender_counts, x="Age", y="count", color="Gender", title="Age distribution",width=700)

fig2.show()

# variante: px.histogram(df, x="Age", color="Gender", title="Age distribution", width=700).show()
```

```python
fig3 = px.box(df, x="Gender", y="Age", title="Age distribution by gender",width=700)
fig3.update_yaxes(dtick=5)

fig3.show()
```

8. What are the different education levels proportions?

```python
counts = (
    df["Education"]
      .value_counts()
      .rename_axis("index ")          # sets name for the category column (with trailing space)
      .reset_index(name="Education")
)

counts
```

```python
px.pie(df, names="Education", title="Education distribution", width=700).show()
```

9. Same question for the education field?

```python
counts = (
    df["EducationField"]
      .value_counts()
      .rename_axis("index ")          # sets name for the category column (with trailing space)
      .reset_index(name="EducationField")
)

counts
```

```python
px.pie(df, names="EducationField", title="Education Fielddistribution", width=700)
```

10. What percentage of the employees never travel for business?

```python
counts = (
    df["BusinessTravel"]
      .value_counts()
      .rename_axis("index ")          # sets name for the category column (with trailing space)
      .reset_index(name="BusinessTravel")
)

print(counts)
```

```python
px.pie(df, names="BusinessTravel", title="Travel", width=700)
```
