Are you a fan of the Game of Thrones series? Let's test it. Let's use the API: <a href="https://anapioficeandfire.com/" target="_blank">https://anapioficeandfire.com/</a> to find out more about the series.

1. Import the following libraries: 

      * ```requests```
      * ```pandas```
      * ```json```

````python
import requests
import pandas as pd

````

2. Go to the URL indicated in the description above. Using `requests`, get data on a character in the series. It doesn't matter who the character is.

````python
#trouver d'abord le endpoint general de l'api
base_url="https://anapioficeandfire.com/api/"
characters_endpoint="characters/"
character_nb='160'
full_url=base_url+characters_endpoint+character_nb
response=requests.get(full_url)
print(response.status_code)
response.content
````

3. Extract these data in JSON format:

````python
response.json()
````

4. Now extract the JSON keys you have obtained and create a DataFrame that will have the JSON keys you extracted as a column.

**Hint**: we can use: `r.json().keys()`.

````python
myKeys=response.json().keys()
myKeys
df=pd.DataFrame([response.json()],columns=myKeys)
df.head()
````

5. Create a loop to extract data from the firsts 150 characters of the series and add them to your DataFrame.

````python
json_data2=[requests.get(base_url+characters_endpoint+str(i)).json() for i in range(1,151)]
json_data2
````

````python
json_data=[]
try:
    for i in range(1,151):
        tmp_characters=requests.get(base_url+characters_endpoint+str(i)).json()
        json_data.append(tmp_characters)
        print(f"Data for character {i} fetched successfully.")
except:
    raise Exception("An error occurred while fetching data from the API.")  
df=pd.DataFrame(json_data)
print(df.shape)
df.head()
````
