```python
import plotly.graph_objects as go
import plotly.express as px
import pandas as pd
iris = px.data.iris()
fig=go.Figure()
fig.add_trace(go.Scatter(x=iris["sepal_width"],y=iris["sepal_length"],mode="markers",marker_color=iris["species_id"]))
fig.update_layout(dict(title="Iris Dataset",xaxis_title="Sepal Width",yaxis_title="Sepal Length"))
fig.show()
```

```python
fig1=px.histogram(iris,x="sepal_width",color="species",title="Distribution of Sepal Width by Species",labels={"sepal_width":"Sepal Width","species":"Species"})
fig1.show()
fig2=px.scatter(iris,x="sepal_width",y="sepal_length",color="species")
fig2.show()
fig2.update_layout(fig1.layout)
```

```python
from plotly.subplots import make_subplots
fig = make_subplots(rows=2, cols=2)
fig.add_trace(fig1.data[0], row=1, col=1)
fig.add_trace(fig1.data[1], row=1, col=2)
fig.add_trace(fig2.data[0], row=2, col=1)
fig.add_trace(fig2.data[1], row=2, col=2)
fig.update_layout(height=400, width=800, title_text="Iris Dataset Visualizations")
fig.show()
print(fig2.data)
```

```python
fig=go.Figure()
fig.add_trace(go.Histogram(x=iris["sepal_width"],name="Sepal Width",visible=True))
fig.add_trace(go.Histogram(x=iris["sepal_length"],name="Sepal length",visible=False))
fig.add_trace(go.Histogram(x=iris["petal_width"],name="Petal Width",visible=False))
fig.add_trace(go.Histogram(x=iris["petal_length"],name="Petal Length",visible=False))

fig.update_layout(updatemenus=[
    dict(
        buttons=[
            go.layout.updatemenu.Button(
                label="Sepal Width",
                method="update",
                args=[{"visible": [True, False, False, False]}, 
                      {"title": {"text": "Distribution of Sepal Width"}}]
            ),
            go.layout.updatemenu.Button(
                label="Sepal Length",
                method="update",
                args=[{"visible": [False, True, False, False]}, 
                      {"title": {"text": "Distribution of Sepal Length"}}]
            ),
            go.layout.updatemenu.Button(
                label="Petal Width",
                method="update",
                args=[{"visible": [False, False, True, False]}, 
                      {"title": {"text": "Distribution of Petal Width"}}]
            ),
            go.layout.updatemenu.Button(
                label="Petal Length",
                method="update",
                args=[{"visible": [False, False, False, True]}, 
                      {"title": {"text": "Distribution of Petal Length"}}]
            )
        ]
    )
])
```
